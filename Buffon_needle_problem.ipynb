This is a nice example of a **Monte Carlo method**, which uses **random sampling** to solve mathematical problems.

>Suppose we have a floor made of parallel strips of wood, each the same width, and we drop a needle onto the floor. What is the probability that the needle will lie across a line between two strips?

Buffon’s needle problem was not originally intended for estimating π. It was first proposed by Georges-Louis Leclerc, Comte de Buffon, in 1733, making it a significant milestone in the history of mathematics. century as a problem in [geometric probability](https://en.wikipedia.org/wiki/Geometric_probability). The problem aimed to determine the probability that a needle dropped randomly on a grid of parallel lines would cross one of the lines.

>Buffon’s needle problem is one of the earliest examples of using probability theory to solve a geometric problem, making it a significant milestone in the history of mathematics.

The connection between the problem and the value of π was discovered later. This connection allowed the problem to be used as a method for estimating π through experimentation or simulation, and it has become a famous example of how π can show up in unexpected places in mathematics. The seemingly unusual appearance of π in this expression occurs because the underlying probability distribution function for the needle orientation is rotationally symmetric.

1.  Take a flat surface and draw parallel lines on it, spaced at equal distances (let’s call this distance d).
2.  Drop a needle of length l (l ≤ d) onto the surface randomly.
3.  Count the number of times the needle crosses one of the lines.
4.  Repeat this process many times (let’s say N times) and keep track of the number of crossings (let’s call this number C).
5.  The probability of the needle crossing a line is equal to (2 * l) / (π * d).
6.  So, we can estimate π using the formula: π ≈ (2 * l * N) / (C * d)

The more times you drop the needle (i.e., the larger N is), the more accurate your estimate of π will be.

Buffon’s needle problem provides a way to approximate the value of π experimentally. It is an experimental approximation of π.

```python
import random #for generating random numbers
import math #for mathematical functions
import matplotlib.pyplot as plt #for plotting
```

<font face="Arial" size="4" color="black">
    
We define the buffon_needle function that takes three parameters: num_needles (the number of needles to drop), needle_length (the length of each needle), and line_distance (the distance between the parallel lines).

Inside the function, crossings is initialized to 0 to keep track of the number of needles crossing the lines. plt.figure(figsize=(8, 8)) creates a new figure with a specified size of 8x8 inches.

We start a loop that runs num_needles times, representing each needle drop.

We randomly generate the position (x, y) and angle (theta) of the needle. random.uniform(0, line_distance) generates a random float between 0 and line_distance, representing the x and y coordinates of the needle's starting point. random.uniform(0, math.pi) 
generates a random angle between 0 and π radians.

We calculate the endpoints of the needle based on its starting position, length, and angle. math.cos(theta) and math.sin(theta) calculate the cosine and sine of the angle, respectively, which are used to determine the x and y components of the needle's length.

We check if the needle crosses a line. If the x-coordinate of the needle's endpoint (x_end) is less than 0 or greater than line_distance, it means the needle crosses a line. In that case, crossings is incremented by 1, and the needle is plotted in red color using plt.plot([x, x_end], [y, y_end], 'r'). Otherwise, the needle is plotted in blue color using plt.plot([x, x_end], [y, y_end], 'b').

We estimate the value of π using the formula: (2 * needle_length * num_needles) / (crossings * line_distance). It calculates the ratio of the total length of needles dropped to the number of crossings multiplied by the distance between the lines.

The loop plots vertical lines at each integer position from 0 to line_distance using plt.axvline(). The lines are plotted with a dashed linestyle (linestyle='--') and black color (color='k').

We set the x-axis and y-axis limits of the plot to range from 0 to line_distance.

We set the title of the plot, which includes the estimated value of π rounded to 4 decimal places. plt.savefig('buffon_needle_plot.png') saves the plot as an image file named "buffon_needle_plotpng".

plt.show() displays the plot on the screen. The function then returns the estimated value of π. 
</font>

```python
#Defining the buffon_needle function that takes the number of needles, needle length, and line distance as parameters.
def buffon_needle(num_needles, needle_length, line_distance):
    #Initializing variables 
    crossings = 0 #to keep track of the number of needles crossing the lines
    plt.figure(figsize=(8, 8))#creating a new figure with a specified size
    
    for _ in range(num_needles):
        # Starting a loop that runs num_needles times, representing each needle drop.
        
        # Randomly generate the position and angle of the needle
        x = random.uniform(0, line_distance)
        y = random.uniform(0, line_distance)
        theta = random.uniform(0, math.pi)
        
        # Calculate the endpoints of the needle
        x_end = x + needle_length * math.cos(theta)
        y_end = y + needle_length * math.sin(theta)
        
        # Check if the needle crosses a line
        if x_end < 0 or x_end > line_distance:
            crossings += 1
            plt.plot([x, x_end], [y, y_end], 'r')
        else:
            plt.plot([x, x_end], [y, y_end], 'b')
    
    # Estimate pi using the formula
    pi_estimate = (2 * needle_length * num_needles) / (crossings * line_distance)
    
    # Plot vertical lines
    for i in range(line_distance + 1):
        plt.axvline(i, linestyle='--', color='k')
    
    plt.xlim(0, line_distance)
    plt.ylim(0, line_distance)
    plt.title(f"Buffon's Needle Simulation (π estimate: {pi_estimate:.4f})")
    plt.savefig('buffon_needle_plot.png')
    
    plt.show()
    
    
    return pi_estimate
```

<font face="Arial" size="4" color="black">

The main() function is defined, which sets the values for num_needles, needle_length, and line_distance. It calls the buffon_needle function with these parameters and stores the estimated value of π in pi_estimate. Finally, it prints the estimated value of π.
</font>

```python
def main():
    num_needles = 200
    needle_length = 1
    line_distance = 2

    pi_estimate = buffon_needle(num_needles, needle_length, line_distance)
    print(f"Estimated value of π: {pi_estimate}")
```

```python
if __name__ == "__main__":
    main()
```

Now I want to modify the code to increase the number of needles and plot the convergence of the π estimation

```python
def buffon_needle(num_needles, needle_length, line_distance):
    # Initialize the count of needles crossing the lines
    crossings = 0
    
    # Initialize a list to store the pi estimates
    pi_estimates = []
    
    # Run the simulation for the given number of needles
    for i in range(1, num_needles + 1):
        # Generate random position (x) and angle (theta) for each needle
        x = random.uniform(0, line_distance)
        theta = random.uniform(0, math.pi)
        
        # Calculate the x-coordinate of the needle's endpoint
        x_end = x + needle_length * math.cos(theta)
        
        # Check if the needle crosses a line and increment the count if so
        if x_end < 0 or x_end > line_distance:
            crossings += 1
        
        # Estimate the value of pi using the formula
        pi_estimate = (2 * needle_length * i) / (crossings * line_distance)
        
        # Append the estimated pi value to the list
        pi_estimates.append(pi_estimate)
    
    # Return the list of pi estimates
    return pi_estimates
```

```python
def main():
    # Set the number of needles, needle length, and line distance
    num_needles = 15000
    needle_length = 1
    line_distance = 2

    # Call the buffon_needle function to get the list of pi estimates
    pi_estimates = buffon_needle(num_needles, needle_length, line_distance)
    
    # Create a plot of pi estimation over the number of needles
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, num_needles + 1), pi_estimates)
    plt.axhline(y=math.pi, color='r', linestyle='--', label='True π')
    plt.xlabel('Number of Needles')
    plt.ylabel('Estimated π')
    plt.title('π Estimation over Number of Needles')
    plt.legend()
    plt.grid(True)
    # Save the plot to a file
    plt.savefig('buffon_needle_plot_pi_convergence_2.png')
    plt.show()
    
    # Print the final estimated value of pi and the actual value of pi
    print(f"Final estimated value of π: {pi_estimates[-1]}")
    print(f"Actual value of π: {math.pi}")

if __name__ == "__main__":
    main()
```
