# Turtle Graphics Library

```python
from turtle import Turtle, Screen
from random import random
from math import sqrt
from time import time
```

<font face="Arial" size="4" color="black">

</font>

```python
# Create a Screen object for the turtle graphics
screen = Screen()
screen.title('Monte Carlo method applied to approximating the value of π')
```

<font face="Arial" size="4" color="black">

We create a Turtle object named t for drawing and configuring its properties:
-  Hiding the turtle cursor using t.hideturtle().
-  Setting the turtle speed to the fastest using t.speed('fastest').
-  Lifting the pen up using t.penup() so the turtle doesn't draw lines between points.
</font>

```python
# Create a Turtle object for drawing
t = Turtle()
t.hideturtle()  # Hide the turtle cursor
t.speed('fastest')  # Set the turtle speed to the fastest
t.penup()  # Lift the pen up so the turtle doesn't draw lines between points
```

<font face="Arial" size="4" color="black">
    
Defining the pi_constant function that takes the number of points n as input:

-   Initializing a variable n_internal to keep track of the count of points inside the unit circle.
-   Starting a loop that iterates n times:
       - Generating random x and y coordinates between 0 and 1 using random().
       - Calculating the distance from the origin to the point using the Pythagorean theorem: sqrt(x**2 + y**2).
       -  Checking if the distance is less than 1 (inside the unit circle):
           - If true, incrementing n_internal and setting the color to "blue".
           - If false, setting the color to "red".
       - Calling the draw_dots function to draw the point at the scaled position with the specified color.
-    Returning the approximation of π by calculating 4 * n_internal / n.
</font>

```python
def pi_constant(n):
    n_internal = 0  # Initialize the count of points inside the circle
    
    for i in range(n):
        x, y = random(), random()  # Generate random x and y coordinates between 0 and 1
        
        distance = sqrt(x**2 + y**2)  # Calculate the distance from the origin to the point
        
        if distance < 1:
            n_internal += 1  # Increment the count if the point is inside the unit circle
            color = "blue"  # Set the color to blue for points inside the circle
        else:
            color = "red"  # Set the color to red for points outside the circle
        
        draw_dots(x, y, color)  # Call the draw_dots function to draw the point
    
    return 4 * n_internal / n  # Calculate and return the approximation of π
```

<font face="Arial" size="4" color="black">

We define the draw_dots function that takes x, y, color, and an optional scale parameter:

  - Moving the tzurtle to the scaled position using t.setposition(x * scale, y * scale).

  - Drawing a dot with the specified color and size using t.dot(4, color).

</font>

```python
def draw_dots(x, y, color, scale=400):
    t.setposition(x * scale, y * scale)  # Move the turtle to the scaled position
    t.dot(4, color)  # Draw a dot with the specified color and size
```

<font face="Arial" size="4" color="black">

 - Calculating the elapsed time by subtracting the start time from the current time using elapsed_time = time() - start_time.

 - Getting the turtle screen using cv = t.getscreen().
 - Saving the drawing as an EPS file named "MonteCarlomethodPi.eps" using cv.getcanvas().postscript(file="MonteCarlomethodPi.eps", colormode='color').
 - Starting the main loop of the turtle screen using t.screen.mainloop() to display the graphics window.


pi_const = pi_constant(15)  # Call the pi_constant function with 15000 points. I will call just with 15 so it is faster. You can do it yourself for 15000 if you like to wait. The image of the 15000 points is in the Plot directory.

</font>

```python
start_time = time()  # Record the start time

pi_const = pi_constant(15)  # Call the pi_constant function with 15000 points

print(chr(960), "=", pi_const)  # Print the approximated value of π

elapsed_time = time() - start_time  # Calculate the elapsed time
print("Time(s):", round(elapsed_time, 2))  # Print the elapsed time rounded to 2 decimal places

cv = t.getscreen()  # Get the turtle screen
cv.getcanvas().postscript(file="MonteCarlomethodPi_try.eps", colormode='color')  # Save the drawing as an EPS file

#t.screen.mainloop()  # Start the main loop of the turtle screen
t.screen.exitonclick()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

<font face="Arial" size="4" color="black">

The monte_carlo_pi function implements the Monte Carlo method to estimate the value of π. Let's break down the code:

- The function takes a parameter num_samples, which represents the number of random points to be generated for the Monte Carlo simulation.
- x = np.random.uniform(-1, 1, num_samples) and y = np.random.uniform(-1, 1, num_samples) generate two arrays of random numbers, x and y, each containing num_samples elements. The random numbers are **uniformly distributed** between -1 and 1. These arrays represent the x and y coordinates of the random points.
- inside_circle = np.cumsum(x**2 + y**2 <= 1) calculates a boolean array that indicates whether each point (x, y) lies inside the unit circle. The condition x**2 + y**2 <= 1 checks if the squared distance from the origin to each point is less than or equal to 1 (the radius of the unit circle). The result is a boolean array where True represents points inside the circle and False represents points outside the circle. The np.cumsum() function then calculates the cumulative sum of this boolean array, which gives the count of points inside the circle at each iteration.
- pi_estimates = 4 * inside_circle / np.arange(1, num_samples + 1) calculates the estimates of π for each iteration. The formula 4 * inside_circle / iteration_number is used to estimate π, where inside_circle is the cumulative count of points inside the circle, and iteration_number is the corresponding iteration number (1 to num_samples). The np.arange(1, num_samples + 1) function generates an array of iteration numbers from 1 to num_samples.
- Finally, the function returns three values:
   -    pi_estimates: An array containing the estimates of π for each iteration.
   -    x: The array of x-coordinates of the random points.
   -    y: The array of y-coordinates of the random points.

The returned pi_estimates array provides the estimates of π at each iteration, allowing to observe the convergence of the estimates as the number of samples increases.
</font>

```python
def monte_carlo_pi(num_samples):
    # Generate random x and y coordinates between -1 and 1 
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)

    # Calculate the cumulative sum of points inside the unit circle 
    inside_circle = np.cumsum(x**2 + y**2 <= 1)
    
    # Calculate the π estimates for each iteration
    pi_estimates = 4 * inside_circle / np.arange(1, num_samples + 1)

    return pi_estimates, x, y
```

<font face="Arial" size="4" color="black">

The plot_monte_carlo function is responsible for visualizing the Monte Carlo simulation results using Matplotlib. Let's break down the code:

-  plt.figure(figsize=(6, 6)) creates a new figure with a specified size of 6 inches by 6 inches. This sets up the canvas for the plot.
-  plt.scatter(x, y, c=np.where(x**2 + y**2 <= 1, "blue", "red"), s=5) creates a scatter plot of the random points. The x and y arrays contain the coordinates of the points. The c parameter is used to assign colors to the points based on a condition. In this case, np.where(x**2 + y**2 <= 1, "blue", "red") assigns the color "blue" to points inside the unit circle and "red" to points outside the unit circle. The s parameter sets the size of the markers to 5.
-  plt.xlabel("x") and plt.ylabel("y") set the labels for the x-axis and y-axis, respectively.
-  plt.title("Monte Carlo Estimation of π") sets the title of the plot.
-  plt.xlim(-1, 1) and plt.ylim(-1, 1) set the limits of the x-axis and y-axis to range from -1 to 1. This ensures that the plot is centered around the origin and covers the unit circle.
-  plt.grid(True) adds a grid to the plot for better readability.
-  plt.savefig("monte_carlo_pi.png") saves the plot as an image file named "monte_carlo_pi.png".
-  plt.show() displays the plot on the screen.

This function takes the x and y arrays containing the coordinates of the random points generated by the Monte Carlo simulation. It creates a scatter plot where the points inside the unit circle are colored blue, and the points outside the unit circle are colored red. The plot provides a visual representation of the Monte Carlo simulation, allowing to observe the distribution of points and the approximation of π.

The resulting plot will have the x-axis labeled as "x", the y-axis labeled as "y", and a title "Monte Carlo Estimation of π". The plot will be saved as an image file named "monte_carlo_pi.png" and displayed on the screen.
</font>

```python
def plot_monte_carlo(x, y):
    plt.figure(figsize=(6, 6))
    plt.scatter(x, y, c=np.where(x**2 + y**2 <= 1, "blue", "red"), s=5)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Monte Carlo Estimation of π")
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.grid(True)
    plt.savefig("monte_carlo_pi.png")
    plt.show()
```

<font face="Arial" size="4" color="black">
    
The plot_pi_estimates function is responsible for visualizing the convergence of the π estimates over iterations using Matplotlib. Let's break down the code:

-   plt.figure(figsize=(8, 4)) creates a new figure with a specified size of 8 inches by 4 inches. This sets up the canvas for the plot.
-   plt.plot(pi_estimates) plots the pi_estimates array, which contains the estimates of π at each iteration. The x-axis represents the iteration number, and the y-axis represents the corresponding π estimate.
-   plt.xlabel("Iterations") sets the label for the x-axis to "Iterations".
-   plt.ylabel("π Estimation") sets the label for the y-axis to "π Estimation".
-   plt.title("Convergence of π Estimation") sets the title of the plot to "Convergence of π Estimation".
-   plt.grid(True) adds a grid to the plot for better readability.
-   plt.savefig("pi_estimation_convergence.png") saves the plot as an image file named "pi_estimation_convergence.png".
-   plt.show() displays the plot on the screen.

This function takes the pi_estimates array, which contains the estimates of π at each iteration of the Monte Carlo simulation. It creates a line plot that shows the convergence of the π estimates over iterations. The x-axis represents the iteration number, and the y-axis represents the corresponding π estimate.

The resulting plot will have the x-axis labeled as "Iterations", the y-axis labeled as "π Estimation", and a title "Convergence of π Estimation". The plot will be saved as an image file named "pi_estimation_convergence.png" and displayed on the screen.

By visualizing the convergence of the π estimates, you can observe how the estimates stabilize and approach the true value of π as the number of iterations increases. This plot provides insights into the accuracy and convergence behavior of the Monte Carlo simulation.
</font>

```python
def plot_pi_estimates(pi_estimates):
    plt.figure(figsize=(8, 4))
    plt.plot(pi_estimates)
    plt.xlabel("Iterations")
    plt.ylabel("π Estimation")
    plt.title("Convergence of π Estimation")
    plt.grid(True)
    plt.savefig("pi_estimation_convergence.png")
    plt.show()
```

<font face="Arial" size="4" color="black">

The main function serves as the entry point of the script and orchestrates the execution of the Monte Carlo simulation and the visualization of the results. Let's break down the code:

-  num_samples = 15000 sets the number of random points to be generated for the Monte Carlo simulation.
-  start_time = time.time() records the start time of the simulation.
-  pi_estimates, x, y = monte_carlo_pi(num_samples) calls the monte_carlo_pi function with the specified number of samples (num_samples). This function performs the Monte Carlo simulation and returns the estimates of π at each iteration (pi_estimates), along with the x and y coordinates of the random points (x and y).
-  end_time = time.time() records the end time of the simulation.
-  elapsed_time = end_time - start_time calculates the elapsed time of the simulation by subtracting the start time from the end time.
-  print(f"Final estimated value of π: {pi_estimates[-1]}") prints the final estimated value of π, which is the last element of the pi_estimates array.
-  print(f"Elapsed time: {elapsed_time:.4f} seconds") prints the elapsed time of the simulation in seconds, formatted to 4 decimal places.
-  plot_monte_carlo(x, y) calls the plot_monte_carlo function to visualize the random points generated during the Monte Carlo simulation. It creates a scatter plot showing the points inside and outside the unit circle.
-  plot_pi_estimates(pi_estimates) calls the plot_pi_estimates function to visualize the convergence of the π estimates over iterations. It creates a line plot showing how the estimates of π stabilize as the number of iterations increases.

Finally, the if __name__ == "__main__": block ensures that the main function is executed only when the script is run directly, and not when it is imported as a module.

When you run this script, it will perform the Monte Carlo simulation with 15000 random points, estimate the value of π, and display the final estimated value and the elapsed time. It will also generate two plots: one showing the distribution of random points inside and outside the unit circle, and another showing the convergence of the π estimates over iterations.

The main function ties together all the components of the script, including the Monte Carlo simulation, the visualization of the random points, and the visualization of the π estimate convergence. It provides a clear entry point and orchestrates the execution of the different parts of the script.
</font>

```python
def main():
    num_samples = 15000
    start_time = time.time()
    pi_estimates, x, y = monte_carlo_pi(num_samples)
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Final estimated value of π: {pi_estimates[-1]}")
    print(f"Elapsed time: {elapsed_time:.4f} seconds")
    plot_monte_carlo(x, y)
    plot_pi_estimates(pi_estimates)

if __name__ == "__main__":
    main()
```
