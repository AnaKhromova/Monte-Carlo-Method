# What is the Probability that a Drunken Man will fall off the Cliff in a 1D Random Walk?


To determine the probability that the “drunken man” will fall off a cliff, we need to define the specific scenario and the cliff’s location. Let’s consider a simple example where the cliff is located at a certain distance from the starting point, and the drunken man takes random steps either toward or away from the cliff.

Let’s say the cliff is located at x =-10, and the drunken man starts at x = 0. At each step, he moves either left (towards the cliff) or right (away from the cliff) with equal probability (0.5 or 1/2).

Now we can simulate the random walk and count the number of times the drunken man reaches or crosses the cliff’s location out of many trials. The probability of falling off the cliff can then be estimated as the ratio of the number of times he falls off the cliff to the total number of trials.

The code simulates this 1D random walk scenario and estimates the probability of the drunken man falling off the cliff based on the specified number of steps and the number of trials. The equal probability of moving left or right is implemented using the random.choice([LEFT, RIGHT]) statement, where LEFT and RIGHT are constants representing the two possible directions.

The drunken man starts at position 0, moves left or right with equal probability at each step, and falls off the cliff if his position becomes less than or equal to the cliff position (-10 in this case).

The LEFT and RIGHT represent the possible directions of movement in the random walk. Using constants makes the code more readable and easier to understand. We define a function named random_walk_1d that takes a single parameter num_steps, which represents the number of steps to be taken in the random walk. Inside this function, we initialize a list called positions with a single element 0. This represents the starting position of the random walk. We start a loop that iterates num_steps times. The underscore _ is used as the loop variable since we don't need to use the loop variable inside the loop.Inside the loop, we use random.choice() to randomly select either LEFT or RIGHT as the direction for the current step. This simulates the random decision of moving left or right at each step. Based on the randomly chosen direction, we update the position by appending a new position to the positions list: If the direction is LEFT, we subtract 1 from the last position in the list (positions[-1]) and append the result to the list; If the direction is RIGHT, we add 1 to the last position in the list and append the result. After the loop finishes, the function returns the positions list, which contains all the positions visited during the random walk.

A function random_walk_1d simulates a 1-dimensional random walk. It takes the number of steps as input and returns a list of positions visited during the random walk. At each step, the direction of movement (left or right) is randomly chosen, and the position is updated accordingly.
</font>

```python
# Constants for the directions
LEFT = 'left'
RIGHT = 'right'

def random_walk_1d(num_steps):
    # Initialize the starting position
    positions = [0]
    
    # Perform the random walk for the given number of steps
    for _ in range(num_steps):
        # Randomly choose the direction of the step
        direction = random.choice([LEFT, RIGHT])
        
        # Update the position based on the direction
        if direction == LEFT:
            positions.append(positions[-1] - 1)
        else:
            positions.append(positions[-1] + 1)
    
    # Return the list of positions visited during the random walk
    return positions
```

This code simulates a 1-dimensional random walk with a cliff. It takes the number of steps and the cliff position as input and returns a boolean value indicating whether the man fell off the cliff during the random walk. The random walk continues until the specified number of steps is reached or the man falls off the cliff, whichever occurs first.

We define a function called random_walk_cliff that simulates a 1-dimensional random walk with a cliff. The function takes two parameters: num_steps, which represents the number of steps to be taken in the random walk, and cliff_position, which represents the position of the cliff. Inside the function, we initialize two variables: position, which represents the current position of the man (starting at 0), and fallen_off, a boolean flag that indicates whether the man has fallen off the cliff (initially set to False). We start a loop that iterates num_steps times to perform the random walk. Inside the loop, we randomly choose the direction of the step using random.choice([LEFT, RIGHT]). Based on the chosen direction, we update the position variable by either decrementing it by 1 (if the direction is LEFT) or incrementing it by 1 (if the direction is RIGHT).

After updating the position, we check if the position has reached or crossed the cliff_position. If the position is less than or equal to cliff_position, it means the man has fallen off the cliff. In this case, we set the fallen_off flag to True and break out of the loop using the break statement. Finally, the function returns the fallen_off flag, which indicates whether the man fell off the cliff during the random walk.

The code simulates a 1-dimensional random walk with a cliff. It takes the number of steps and the cliff position as input and returns a boolean value indicating whether the man fell off the cliff during the random walk. The random walk continues until the specified number of steps is reached or the man falls off the cliff, whichever occurs first.
</font>

```python
def random_walk_cliff(num_steps, cliff_position):
    # Initialize the starting position and fallen_off flag
    position = 0
    fallen_off = False
    
    # Perform the random walk for the given number of steps
    for _ in range(num_steps):
        # Randomly choose the direction of the step
        direction = random.choice([LEFT, RIGHT])
        
        # Update the position based on the direction
        if direction == LEFT:
            position -= 1
        else:
            position += 1
        
        # Check if the position has reached or crossed the cliff
        if position <= cliff_position:
            fallen_off = True
            break
    
    # Return the fallen_off flag indicating if the man fell off the cliff
    return fallen_off
```

```python
def main():
    num_steps = 1000
    num_walks = 10
    cliff_position = -10
    num_trials = 100000
    
    plt.figure(figsize=(8, 6))
    
    # Simulate and plot multiple random walks
    for i in range(num_walks):
        positions = random_walk_1d(num_steps)
        plt.plot(positions, label=f'Walk {i+1}')
        
    # Add the cliff position to the plot
    plt.axhline(y=cliff_position, color='r', linestyle='--', label='Cliff')    
    
    plt.xlabel('Number of Steps')
    plt.ylabel('Position')
    plt.title('1D Random Walk')
    plt.legend()
    plt.grid(True)
    plt.savefig('1d_random_walk.png')
    plt.show()
    
    # Calculate the probability of falling off the cliff
    results = [random_walk_cliff(num_steps, cliff_position) for _ in range(num_trials)]
    fallen_off_count = results.count(True)
    probability = fallen_off_count / num_trials    
    print(f"Probability of falling off the cliff: {probability:.4f}")
   
# Run the main function if the script is executed directly
if __name__ == "__main__":
    main()
```

# A 2D Random Walk, also known as the Drunken Man Problem

The random walk simulation in the code uses the random.choice(['up', 'down', 'left', 'right'])function to select a random direction at each walk's step. The available directions are 'up', 'down', 'left', and 'right'. This introduces randomness into the simulation, making it a Monte Carlo method. Monte Carlo methods are characterized by the use of random sampling to solve problems that might be deterministic in principle. In the case of the random walk, the problem is to simulate the path of a "drunken man" who takes random steps in different directions. By using random.choice() function to select the direction at each step, we are essentially sampling from a set of possible outcomes, which is a key feature of Monte Carlo simulations.
We import matplotlib.pyplot as plt and random Python packages as with a 1D random walk case.

Then, we define a function random_walk_2d that simulates a 2D random walk. This function simulates a 2D random walk for a given number of steps (num_steps). It initializes the starting position (x, y) to (0, 0), representing the initial coordinates of the random walk. It initializes an empty list called positions to store the positions visited during the random walk. The starting position (0, 0) is added to the list. It starts a loop that iterates num_steps times to perform the random walk. Inside the loop, it randomly chooses a direction for each step using random.choice(['up', 'down', 'left', 'right']). The available directions are 'up', 'down', 'left', and 'right'. Based on the chosen direction, it updates the x and y coordinates accordingly: If the direction is 'up', it increments y by 1; If the direction is 'down', it decrements y by 1; If the direction is 'left', it decrements x by 1; If the direction is 'right', it increments x by 1. It appends the new position (x, y) to the positions list. After the loop finishes, the function returns the positions list, which contains all the positions visited during the random walk.
</font>

```python
def random_walk_2d(num_steps):
    # Initialize the starting position (x, y)
    x = 0
    y = 0
    
    # Initialize a list to store the positions visited during the random walk
    positions = [(x, y)]
    
    # Perform the random walk for the given number of steps
    for _ in range(num_steps):
        # Randomly choose the direction of the step
        direction = random.choice(['up', 'down', 'left', 'right'])
        
        # Update the x and y coordinates based on the chosen direction
        if direction == 'up':
            y += 1
        elif direction == 'down':
            y -= 1
        elif direction == 'left':
            x -= 1
        else:
            x += 1
        
        # Append the new position to the list of positions
        positions.append((x, y))
    
    # Return the list of positions visited during the random walk
    return positions
```

This function simulates a 2D random walk by randomly choosing a direction at each step and updating the coordinates accordingly. The resulting list of positions represents the path taken during the random walk.

-   It plots the random walk using plt.plot(x, y, label=f'Walk {i+1}'), where label is used to assign a label to each random walk.

-   Finally, it displays the plot using plt.show().

```python
def main():
    # Set the number of steps and the number of random walks to simulate
    num_steps = 1000
    num_walks = 10
    
    # Create a new figure with a specified size
    plt.figure(figsize=(8, 8))
    
    # Simulate and plot multiple random walks
    for i in range(num_walks):
        # Perform a 2D random walk and get the list of positions
        positions = random_walk_2d(num_steps)
        
        # Unpack the list of positions into separate lists of x and y coordinates
        x, y = zip(*positions)
        
        # Plot the random walk
        plt.plot(x, y, label=f'Walk {i+1}')
    
    # Set the labels for the x-axis and y-axis
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    
    # Set the title of the plot
    plt.title('2D Random Walk')
    
    # Add a legend to the plot
    plt.legend()
    
    # Enable the grid for better readability
    plt.grid(True)
    
    # Save the plot to a file named '2d_random_walk.png'
    plt.savefig('2d_random_walk.png')
    
    # Display the plot
    plt.show()

# Run the main function if the script is executed directly
if __name__ == "__main__":
    main()
```

# What is the Probability that a Drunken Man will fall off the Cliff in a 2D Random Walk?

The problem of determining the probability of a "drunken man" falling off a cliff in a 2D random walk is a classic example in the field of stochastic processes and probability theory. It involves simulating the random movement of an object or person in a two-dimensional space and analyzing the likelihood of reaching a specific boundary or condition.

The goal is to determine the probability that the "drunken man" will fall off the cliff after a certain number of steps. This probability can be estimated by running multiple simulations of the random walk and counting the number of times the man falls off the cliff out of the total number of simulations.

The problem can be approached using the Monte Carlo simulation, which involves generating random samples and observing the outcomes. By simulating a large number of random walks, we can approximate the probability of falling off the cliff.

The import is the same. The full version of the code is in the Jupyter Notebook in a GitHub repository here. The code simulates a single instance of the 2D random walk and checks if the man falls off the cliff based on the specified cliff boundaries. To estimate the probability, you would need to run this simulation multiple times and calculate the ratio of simulations where the man falls off the cliff to the total number of simulations. It also highlights the concept of boundary conditions and the importance of defining clear stopping criteria in stochastic processes.

The random_walk_2d function takes three parameters: num_steps - the number of steps to perform in the random walk; cliff_x - the x-coordinate of the cliff boundary; cliff_y -  the y-coordinate of the cliff boundary. The function starts by initializing the starting position (x, y) to (0, 0) and creates an empty list called positions to store the positions visited during the random walk. It also initializes a flag fallen_off to track if the drunken man falls off the cliff.
The random walk is performed for the given number of steps using a loop. In each iteration. A random direction is chosen from the list ['up', 'down', 'left', 'right'] using random.choice(). The x and y coordinates are updated based on the chosen direction. If the direction is 'up', y is incremented by 1. If the direction is 'down', y is decremented by 1. If the direction is 'left', x is decremented by 1. If the direction is 'right', x is incremented by 1. The new position (x, y) is appended to the positions list. The code checks if the current position has reached or crossed the cliff boundary by comparing x with cliff_x and y with cliff_y. If either condition is true, the fallen_off flag is set to True, and the loop is terminated using the break statement. Finally, the function returns the list of positions visited (positions) and the fallen_off flag.



</font>

```python
def random_walk_2d(num_steps, cliff_x, cliff_y):
    # Initialize the starting position (x, y)
    x = 0
    y = 0
    
    # Initialize a list to store the positions visited during the random walk
    positions = [(x, y)]
    
    # Initialize a flag to track if the drunken man falls off the cliff
    fallen_off = False
    
    # Perform the random walk for the given number of steps
    for _ in range(num_steps):
        # Randomly choose the direction of the step
        direction = random.choice(['up', 'down', 'left', 'right'])
        
        # Update the x and y coordinates based on the chosen direction
        if direction == 'up':
            y += 1
        elif direction == 'down':
            y -= 1
        elif direction == 'left':
            x -= 1
        else:
            x += 1
        
        # Append the new position to the list of positions
        positions.append((x, y))
        
        # Check if the current position has reached or crossed the cliff boundary
        if x <= cliff_x or y <= cliff_y:
            fallen_off = True
            break
    
    # Return the list of positions visited and the fallen_off flag
    return positions, fallen_off
```

The main function sets up the simulation parameters,  performs multiple random walks, and calculates the probability of  falling off the cliff. the main function, sets up the simulation parameters, performs multiple random walks using the random_walk_2d function, and calculates the probability of falling off the cliff based on the simulation results. By running multiple simulations (num_walks times) and counting the number of times the man falls off the cliff (fallen_off_count), the code estimates the probability of falling off the cliff as the ratio of fallen_off_count to num_walks.

In the main function:

-   It sets the number of steps (num_steps) to 1000 and the number of walks (num_walks) to 10000.
-   It initializes a counter (fallen_off_count) to keep track of the number of walks where the man falls off the cliff.
-   It runs a loop that iterates num_walks times. In each iteration, it calls the random_walk_2d function to perform a single random walk and checks if the man falls off the cliff.
-   If the man falls off the cliff during a walk, it increments the fallen_off_count.
-   After all the walks are simulated, it calculates the probability of falling off the cliff by dividing the fallen_off_count by the total number of walks (num_walks).

By running multiple simulations (num_walks times) and counting the number of times the man falls off the cliff (fallen_off_count), the code estimates the probability of falling off the cliff as the ratio of fallen_off_count to num_walks.



</font>

```python
def main():
    # Set the number of steps, number of walks, and cliff boundary coordinates
    num_steps = 1000
    num_walks = 10000
    cliff_x = -10
    cliff_y = -10
    
    # Initialize a counter for the number of walks where the drunken man falls off the cliff
    fallen_off_count = 0
    
    # Simulate multiple random walks
    for _ in range(num_walks):
        # Perform a 2D random walk and get the list of positions and fallen_off flag
        positions, fallen_off = random_walk_2d(num_steps, cliff_x, cliff_y)
        
        # Increment the fallen_off_count if the drunken man falls off the cliff
        if fallen_off:
            fallen_off_count += 1
    
    # Calculate the probability of falling off the cliff
    probability = fallen_off_count / num_walks
    
    # Print the probability of falling off the cliff
    print(f"Probability of falling off the cliff 2D: {probability:.4f}")

# Run the main function if the script is executed directly
if __name__ == "__main__":
    main()
```

# Could the Drunken Man return to the Starting Point?

```python
from scipy.special import comb

def probability_at_origin_1d(steps):
    n = steps // 2
    prob = comb(2*n, n, exact=True) / 2**(2*n)
    return prob

# Example usage
steps = 1000
result = probability_at_origin_1d(steps)
print(f"The probability of being at the origin after {steps} steps in a 1D random walk is: {result}")
```

```python
from scipy.special import comb
from fractions import Fraction

def probability_at_origin_2d(steps):
    n = steps // 2
    prob = Fraction(0)
    for k in range(n + 1):
        term1 = Fraction(1, 4) ** (2 * n)
        term2 = Fraction(comb(2 * n, n, exact=True) * comb(n, k, exact=True) ** 2, 1)
        prob += term1 * term2
    return prob

# Example usage
steps = 1000
result = probability_at_origin_2d(steps)
print(f"The probability of being at the origin after {steps} steps in a 2D random walk is: {float(result)}")
```

## - The 2D Random Walk using the Monte Carlo Simulation

The Monte Carlo method is a numerical approach that involves simulating a large number of random walks and estimating the probability based on the fraction of walks that return to the starting point. As the number of simulated walks increases, the estimated probability converges to the true probability. We use the method to find the probability of returning to the starting point using the following code.

We define the random_walk_2d function that takes the number of steps as input and performs a 2D random walk. It initializes the starting position (x, y) as (0, 0) and stores the positions visited during the random walk in a list called positions. Inside the random_walk_2d function, we used a loop to perform the random walk for the given number of steps. At each step, we randomly choose a direction (up, down, left, or right) and update the x and y coordinates accordingly. The new position is appended to the positions list.

We defined the main function, which sets the number of steps (num_steps) to 1000 and the number of walks (num_walks) to 100000. Inside the main function, we initialized a counter return_to_start_count to keep track of the number of walks that return to the starting point. We used a loop to simulate multiple random walks. For each walk, we called the random_walk_2d function to perform a 2D random walk and stored the list of positions. After each walk, we check if the final position (positions[-1]) is equal to the starting point (0, 0). If it is, you incremented the return_to_start_count. After simulating all the walks, we can calculate the probability of returning to the starting point by dividing return_to_start_count by num_walks. Finally, we print the calculated probability with 10 decimal places.
</font>

```python
import random
import matplotlib.pyplot as plt

def random_walk_2d(num_steps):
    # Initialize the starting position (x, y)
    x = 0
    y = 0
    
    # Initialize a list to store the positions visited during the random walk
    positions = [(x, y)]
    
    # Perform the random walk for the given number of steps
    for _ in range(num_steps):
        # Randomly choose the direction of the step
        direction = random.choice(['up', 'down', 'left', 'right'])
        
        # Update the x and y coordinates based on the chosen direction
        if direction == 'up':
            y += 1
        elif direction == 'down':
            y -= 1
        elif direction == 'left':
            x -= 1
        else:
            x += 1
        
        # Append the new position to the list of positions
        positions.append((x, y))
    
    # Return the list of positions visited
    return positions

def main():
    # Set the number of steps and number of walks
    num_steps = 1000
    num_walks = 100000
    
    # Initialize a counter for the number of walks returning to the starting point
    return_to_start_count = 0
    
    # Simulate multiple random walks
    for _ in range(num_walks):
        # Perform a 2D random walk and get the list of positions
        positions = random_walk_2d(num_steps)
        
        # Check if the final position is the starting point (0, 0)
        if positions[-1] == (0, 0):
            return_to_start_count += 1
    
    # Calculate the probability of returning to the starting point
    probability = return_to_start_count / num_walks
    
    # Print the probability of returning to the starting point
    print(f"Probability of returning to the starting point: {probability:.10f}")

# Run the main function if the script is executed directly
if __name__ == "__main__":
    main()
```
